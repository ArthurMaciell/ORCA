```python
data = pd.read_excel('artifacts\data_ingestion\itens.xlsx')
data.head()
```

```python
# 2. Opcional: remover colunas inúteis como 'Unnamed: 0'
df = data.loc[:, ~data.columns.str.contains('^Unnamed')]

# 3. Converter os tipos para YAML
def map_dtype(dtype):
    if pd.api.types.is_float_dtype(dtype):
        return "float"
    elif pd.api.types.is_integer_dtype(dtype):
        return "int"
    elif pd.api.types.is_bool_dtype(dtype):
        return "bool"
    else:
        return "object"

schema = {
    "columns": {
        col: map_dtype(dtype)
        for col, dtype in df.dtypes.items()
    }
}

# 4. Salvar como schema.yaml
with open("schema.yaml", "w") as f:
    yaml.dump(schema, f, allow_unicode=True)

print("✅ schema.yaml gerado com sucesso!")
```
