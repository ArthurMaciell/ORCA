```python
def prep_ar_sem_peso_volume(df):
    df_prep = df[df['Modelo'] == 'AR'].copy()
    # Transformando a Coluna data em tipo Datatime
    df_prep['Data'] = pd.to_datetime(df_prep['Data'], format="%d/%m/%Y")
    
    # Calcular área e criando nova variável
    df_prep["Area"] = df_prep["Comprimento"] * df_prep["Largura"]
    
    # Substituir valores inválidos por NaN
    df_prep.replace({"Area": {np.inf: np.nan, -np.inf: np.nan}}, inplace=True)
    
    # Criar a flag indicando que a área está ausente
    df_prep["AreaAusente"] = df_prep["Area"].isna()
    
    # Preencher os NaNs da área com 0
    df_prep["Area"] = df_prep["Area"].fillna(0)
    
    # Convertendo Preço unitário em número
    df_prep['Preço Unitário'] = pd.to_numeric(df_prep['Preço Unitário'], errors='coerce')
    
    # Convertendo Fator em número
    df_prep['Fator'] = pd.to_numeric(df_prep['Fator'], errors='coerce')
    
    # Criando nova variável com o preço com fator 1
    #df_prep['Preço (1)'] = df_prep['Preço Unitário']/df_prep['Fator']
    
    #df_prep['Preço/Área'] = df_prep['Preço (1)']/df_prep['Area']
    
    #df_prep = df_prep[df_prep['Data'].dt.year != 2022]
    #df_prep = df_prep[df_prep['Data'].dt.year != 2023]
    
    # Multiplicando o valor de VAT, ADLQ em 2022 por 1.20 (Foi o aumento dessa data até agora)
    df_prep.loc[(df_prep['Data'].dt.year == 2022) & (df_prep['Modelo'].isin(['AR'])), 'Preço (1)'] *= 1.2
    df_prep.loc[(df_prep['Data'].dt.year == 2023) & (df_prep['Modelo'].isin(['AR'])), 'Preço (1)'] *= 1.10
    df_prep.loc[(df_prep['Data'].dt.year == 2024) & (df_prep['Modelo'].isin(['AR'])), 'Preço (1)'] *= 1.15
    #df_prep.loc[df_prep['Largura'] > df_prep['Comprimento'], 'Preço (1)'] *= 1.15

    #Tirando a primeira Coluna

    def categorize_price(price):
        if price <= q1:
            return 'Low'
        elif price <= q2:
            return 'Medium'
        else:
            return 'High'
    
    #Apply the categorization
    df_prep['preço_categoria'] = df_prep['Preço (1)'].apply(categorize_price)

    #df_prep["Preço/Area"] = df_prep["Preço (1)"] / df_prep["Area"].replace(0, np.nan)
    df_prep["Comprimento/Largura"] = df_prep["Comprimento"] / df_prep["Largura"].replace(0, np.nan)

    # Preenche NaNs resultantes com 0
    #df.fillna({"Peso/Área": 0, "Preço/Volume": 0, "Comprimento/Largura": 0}, inplace=True)

    #df_prep["Preço/Area"] = df_prep["Preço/Area"].fillna(0)
    df_prep["Comprimento/Largura"] = df_prep["Comprimento/Largura"].fillna(0)


    primeira_coluna = df_prep.columns[0]
    df_prep = df_prep.drop(columns=[primeira_coluna])
    # Tirando as linhas que não tem modelo preenchido
    df_prep = df_prep[df_prep['Modelo'].notna() & (df_prep['Modelo']!='')]
    
    colunas_vazias = df_prep.columns[df_prep.isna().all()].tolist()
    colunas_tirar = ['Descricao','Linha', 'Data','Qnt','Fator', 'Total','Peso','Volume', 'Preço Unitário']
    df_prep = df_prep.drop(columns=colunas_vazias)
    df_prep = df_prep.drop(columns=colunas_tirar)
    colunas_numericas = ['Comprimento','Largura' ]
    colunas_categoricas = df_prep.select_dtypes(include='float64').columns
    
    # Preenchendo com 0 as variáveis númericas que estavam NaN
    df_prep[colunas_numericas] = df_prep[colunas_numericas].fillna(0)
    
    # Selecionando as colunas categóricas
    colunas_categoricas = df_prep.select_dtypes(include='object').columns
    
    # Preenchendo valores NaN com Ausente nas variáveis categóricas
    df_prep[colunas_categoricas] = df_prep[colunas_categoricas].fillna("Ausente")
    df_prep['Moldura'] = df_prep['Moldura'].fillna("Ausente")
    
    frequencia = df_prep['Modelo'].value_counts()
    #modelos = frequencia[frequencia>200]
    #df_curto = df_prep[df_prep['Modelo'].isin(modelos.index)].copy()
    df_curto = df_prep.copy()
    
    #df_curto = df_curto[df_curto['Preço (1)'] <= 1000]
    
    df_encoded = pd.get_dummies(df_curto)
    
    duplicadas = df_encoded[df_encoded.duplicated()]
    df_encoded_ar = df_encoded.drop_duplicates()
    
    return df_prep, df_encoded_ar , colunas_numericas, colunas_categoricas
```

```python
from dataclasses import dataclass
from pathlib import Path
```

```python
@dataclass(frozen=True)
class DataValidationConfig:
    root_dir: Path
    status_file: str
    data_file_path: Path
    all_schema: dict
```

# === CONFIGURATION MANAGER ===

```python
from ORCA.constants import *
from ORCA.utils.common import read_yaml, create_directories, write_text, is_dtype_compatible
```

```python
class ConfigurationManager:
    def __init__(
        self,
        config_filepath = CONFIG_FILE_PATH,
        params_filepath = PARAMS_FILE_PATH,
        schema_filepath = SCHEMA_FILE_PATH):

        self.config = read_yaml(config_filepath)
        self.params = read_yaml(params_filepath)
        self.schema = read_yaml(schema_filepath)

        create_directories([self.config.artifacts_root])
    
    def get_data_validation_config(self) -> DataValidationConfig:
        config = self.config.data_validation
        schema = self.schema.COLUMNS

        create_directories([config.root_dir])

        data_validation_config = DataValidationConfig(
            root_dir=config.root_dir,
            status_file=config.status_file,
            data_file_path = config.data_file_path,
            all_schema=schema,
        )

        return data_validation_config
```

# === COMPONENT ===

```python
import os
from ORCA import logger
```

```python
class DataValidation:
    def __init__(self, config: DataValidationConfig):
        self.config = config


    def validate_all_columns(self)-> bool:
        try:
            validation_status = None

            data = pd.read_excel(self.config.data_file_path)
            all_cols = list(data.columns)

            all_schema = self.config.all_schema.keys()

            
            for col in all_cols:
                if col not in all_schema:
                    validation_status = False
                    with open(self.config.status_file, 'w') as f:
                        f.write(f"Validation status: {validation_status}")
                else:
                    validation_status = True
                    with open(self.config.status_file, 'w') as f:
                        f.write(f"Validation status: {validation_status}")

            return validation_status
        
        except Exception as e:
            raise e
        

```

```python
try:
    config = ConfigurationManager()
    data_validation_config = config.get_data_validation_config()
    data_validation = DataValidation(config=data_validation_config)
    data_validation.validate_all_columns()
except Exception as e:
    raise e
```
